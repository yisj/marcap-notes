```python
TARGET_DATE = "2021-11-5"
```

```python
# 기본 모듈 로드
import pandas as pd
from marcap import marcap_data
import matplotlib.pyplot as plt
```

```python
basic = pd.read_csv("marcap/financial_statement/basic-211102.csv", dtype={"Code": object})
profit = pd.read_csv("marcap/financial_statement/profit-211102.csv", dtype={"Code": object})
```

```python
basic
```

```python
profit
```

```python
m = marcap_data(TARGET_DATE)
m
```

```python
codes = basic['Code'].unique()
len(codes)
```

```python
data = list()
for code in codes:
    mc = m[m['Code'] == code]
    bc = basic[basic['Code'] == code]
    pc = profit[profit['Code'] == code]
    try:
        data.append([code, mc['Name'].iloc[0], mc['Close'].iloc[0], mc['Stocks'].iloc[0], mc['Marcap'].iloc[0], bc['Assets'].iloc[0], bc['Liabilities'].iloc[0], bc['Equity'].iloc[0], pc['Profit'].iloc[0]])
    except Exception as e:
        print(e, code)
        pass
```

```python
data[0]
```

```python
df = pd.DataFrame(data, columns = ['Code', 'Name', 'Close', 'Stocks', 'Marcap', 'Assets', 'Liabilities', 'Equity', 'Profit'])
df
```

```python
df['P/E'] = df["Marcap"] / df["Equity"]
df
```

```python
df.sort_values(['P/E'], ascending=[True]).to_csv('PE-211105.csv')
```
