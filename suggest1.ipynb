```python
wuryang = df[df['Dept']=='우량기업부']
wuryang.reset_index(inplace=True)
wuryang
```

```python
w2011 = wuryang[wuryang['Date']=='2011-05-02']
```

```python
w2011['Current'] = 0
for index, row in w2011.iterrows():
    w2011['Current'][index] = df[df['Code']==row['Code']].iloc[-1]['Close']
```

```python
# 우량기업부 중에서, 날짜가 가장 최근인 것은?
w_latest = wuryang.groupby('Code')['Date'].min()
```

```python
w_latest
```

```python
w_latest.unique()
```

```python
w_latest_codes = w_latest[w_latest == '2020-05-04']
```

```python
codes = w_latest_codes.index
```

```python
codes
```

```python
data=list()
df.reset_index(inplace=True)
for code in codes:
    c = df[df['Code']==code]
    latest_stocks = c['Stocks'].iloc[-1]
    c['Adj Close'] =  c['Close'] * c['Stocks'] / latest_stocks
    f = c.iloc[0]
    l = c.iloc[-1]
    data.append([f['Name'],f['Date'], f['Adj Close'], l['Date'], l['Adj Close'], l['Adj Close']/f['Adj Close']])

column_names = ["Name", "FDate", "First", "LDate", "Last", "L/F"]

w = pd.DataFrame(data, columns=column_names)
```
