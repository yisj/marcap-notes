```python
# 기본 모듈 로드
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
from marcap import marcap_data
import matplotlib.pyplot as plt
# 코드 검색하기
target_name = "LG화학"
df = pd.read_csv("marcap/data/marcap-2021.csv.gz")
target_code = df[df["Name"] == target_name]["Code"].iloc[0]
target_code
```

```python
# 그래프 설정
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y')

plt.rcParams["figure.figsize"] = (16,4)
plt.rcParams['lines.linewidth'] = 0.5
plt.rcParams["axes.grid"] = True
plt.rcParams["axes.formatter.useoffset"] = False
plt.rcParams["xtick.minor.visible"] = True 
```

```python
# 전체 데이터 불러오기
df = marcap_data('1995-5-2', '2021-12-31', target_code)
```

```python
latest_stocks = df['Stocks'].iloc[-1]
df['Adj Close'] = df['Close'] * df['Stocks'] / latest_stocks
latest_stocks
```

```python
pd.options.display.float_format = '{:.1f}'.format
df['Adj Close'].plot()
```
