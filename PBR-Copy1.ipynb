```python
TARGET_DATE = "2021-11-5"
```

```python
# 기본 모듈 로드
import pandas as pd
from marcap import marcap_data
import matplotlib.pyplot as plt
```

```python
basic = pd.read_csv("marcap/financial_statement/basic.csv", dtype={"code": object})
profit = pd.read_csv("marcap/financial_statement/profit.csv", dtype={"Code": object})
profit
```

```python
m = marcap_data(TARGET_DATE)
m
```

```python
codes = basic['code'].unique()
len(codes)
```

```python
data = list()
for code in codes:
    mc = m[m['Code'] == code]
    bc = basic[basic['code'] == code]
    pc = profit[profit['Code'] == code]
    try:
        data.append([code, mc['Name'].iloc[0], mc['Marcap'].iloc[0], bc['assets'].iloc[0], bc['liabilities'].iloc[0], bc['equity'].iloc[0], pc['Profit'].iloc[0]])
    except:
        pass
```

```python
df = pd.DataFrame(data, columns = ['Code', 'Name', 'Marcap', 'Assets', 'Liabilities', 'Equity', 'Profit'])
df
```

```python
df['P/E'] = df["Marcap"] / df["Equity"]
df
```

```python
df.sort_values(['P/E'], ascending=[True]).to_csv('PE.csv')
```
